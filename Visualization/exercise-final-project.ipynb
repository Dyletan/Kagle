```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Fill in the line below: Specify the path of the CSV file to read
bank_filepath = r"C:\Self Learning\Data Science\Kaggle\Visualization\bank.csv"
```

### Load the data

```python

bank_data = pd.read_csv(bank_filepath)
random_sample = bank_data.sample(n=2000, random_state=42)
```

```python
random_sample.head()
```

### Visualize the data

```python
print(list(bank_data.columns))
plt.figure(figsize=(16,10))
sns.set_style('whitegrid')
sns.swarmplot(x='deposit', y='campaign', hue='loan', data=random_sample)
```
